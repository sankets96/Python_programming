**Using context manager (with)**

*   it good to close file after usage as it will free up memory
*   if we dont close it garbage collector would be close it

*   use of with keyword u dont required to close function






```python
with open('sample3.txt','w') as f:
  f.write('snkeeetttt')
```

```python
#read with keyword
with open('sample2.txt','r') as f:
  print(f.readline())
```

```python
#read 10 char again 10 char
with open('sample1.txt','r') as f:
  print(f.read(10))
  print(f.read(10))
```

```python
#seek and tell function
with open('sample1.txt','r') as f:
  print(f.read(10))
  print(f.tell())  #cursor is in 10 position
```

```python
#seek fn
with open('sample1.txt','r') as f:
  print(f.read(10))
  print(f.tell())
  f.seek(5)              #move cursor to position 5
  print(f.readline())
```

```python
#seek in wirite
with open('sample1.txt','w') as f:
  f.write('sanket')
  f.seek(0)
  f.write('X')   #only x replace ->Xanket
```

**Serialization and Deserialization**

*   Serialization : converting python datatypes into Json format
*   Deserialization : converting JSON to python datatypes



```python
import json
```

```python
#list dump into json module
L=[1,2,3,4,5]
with open('demo.json','w') as f:
  json.dump(L,f)
```

```python
#dic with json module
di ={'name':'sanket','age':'22','ss':'pp'}
with open('demo1.json','w') as f:
  json.dump(di,f)
```

```python
#read the data from json file
with open('demo1.json','r') as f:
  d = json.load(f)
  print(d)
```

```python
#with tuples
t = (12,13,13,14,15)
with open('2.json','w') as f:
  json.dump(t,f)
  #it store into list not in tuple
```

**Serialization and Deserialization with custom obj**

```python
class person:
  def __init__(self,name,surname,age):
    self.name= name
    self.surname= surname
    self.age= age

p1 = person('sanket','suryavanshi','12')
```

```python
def print_p(person):
  if isinstance(person,person):
    return 'Name => {} Surname => {} Age => {}'.format(person.name,person.surname,person.age)
```

```python
with open('new.json','w') as f:
  json.dump(person,f,default=print_p)
```

**Pickling**

*   is the process whereby a Python object hierarchy is converted into a byte stream
*   convert into binary and used into other file and that file extract obj from that pickle



```python
class person:
  def __init__(self,name,surname,age):
    self.name= name
    self.surname= surname
    self.age= age

  def info(self):
    return 'Name =>{} surname => {},age => {}'.format(self.name,self.surname,self.age)


p = person('sankety','sury','33')
```

```python
import pickle
```

```python
with open('pick.pkl','wb') as f:
  pickle.dump(p,f)
```

```python
#retrive
with open('pick.pkl','rb') as f:
  d =pickle.load(f)
```

```python
d.info()
```

**where to use pickle and where to use json**

*   Pickle : lets user to store data into binary format
*   Json : lets user store data in a human-redable text format
